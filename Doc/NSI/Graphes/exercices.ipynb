```python
class Graphe:    
    def __init__(self):
        self.d = dict() # dictionnaire d'adjacence
        
    def ajoute_sommet(self,s): 
        self.d[s] = list()
    
    def ajoute_arete(self,s,t):
        if s not in self.d :
            self.ajoute_sommet(s)
        if t not in self.d :
            self.ajoute_sommet(t)

        self.d[s].append(t)
        self.d[t].append(s)
        
    def arete(self,s,t):
        for element in self.d[s] :
            if element == t :
                return True
        return False
    
    def sommets(self):
        sommet = list()
        for element in self.d :
            sommet.append(element)
        return sommet

    
    def voisins(self,s):
        return self.d[s]
    
    def degre(self,s):
        return len(self.d[s])
    
    def nb_aretes(self):
        cpt = 0
        for element in self.d :
            for _ in self.d[element] :
                cpt += 1
        return cpt /2
    
    def vide(self) :
        self.d = dict()
```

```python
G2 = Graphe()
```

```python
l1 = ["A","B","C","D"]

for element in l1 :
    G2.ajoute_sommet(element)

G2.sommets()
```

```python
G2.ajoute_arete("A","B")
G2.ajoute_arete("A","D")
G2.ajoute_arete("B","C")
G2.ajoute_arete("B","D")

G2.voisins("A")

```

```python
def myst(G,s):
    noir = []
    pile = [s]
    while len(pile) > 0 :
        u = pile.pop()
        if u not in noir :
            noir.append(u)
            for v in G[u]:
                pile.append(v)  
    return noir
```
