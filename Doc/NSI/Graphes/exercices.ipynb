```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph() # cr√©e un graphe vide
G.add_nodes_from(["A","B","C","D","E","F"])
G.add_edges_from([("A", "B"),("A", "C"),("A", "D"),("B", "E"),("B", "F"),("E", "F")])
```

```python
plt.clf() # on efface
nx.draw(G, with_labels=True) # on dessine
plt.show()
```
